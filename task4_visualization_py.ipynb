```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def create_visuals():
    # --- 1. LOAD THE SUMMARY DATA ---
    try:
        # Load the stats we created in Task 3
        summary_df = pd.read_csv('data/category_summary.csv')

        # We set 'category' as the index so it shows up on the labels of our chart
        summary_df.set_index('category', inplace=True)
    except FileNotFoundError:
        print("Error: 'data/category_summary.csv' not found. Please run Task 3 first!")
        return

    # --- 2. CREATE THE BAR CHART ---
    print("Generating chart...")

    # We create a figure (the window) and an axis (the actual chart)
    plt.figure(figsize=(10, 6))

    # We create a bar chart of the average scores per category
    # 'color' can be a name or a hex code
    summary_df['avg_score'].plot(kind='bar', color='skyblue', edgecolor='black')

    # --- 3. CUSTOMIZE THE CHART ---
    # Adding titles and labels makes the data professional
    plt.title('Average Story Popularity by Category (TrendPulse)', fontsize=14)
    plt.xlabel('Category', fontsize=12)
    plt.ylabel('Average Upvote Score', fontsize=12)
    plt.xticks(rotation=45) # Rotate category names so they don't overlap
    plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a light grid for readability

    # --- 4. SAVE THE VISUAL ---
    # We ensure the data folder exists (just in case)
    if not os.path.exists('data'):
        os.makedirs('data')

    output_path = 'data/category_popularity.png'
    plt.tight_layout() # This prevents the labels from getting cut off
    plt.savefig(output_path)

    print(f"Chart successfully saved to {output_path}")

    # This opens a window to show you the chart immediately
    plt.show()

if __name__ == "__main__":
    create_visuals()
```

```python
import pandas as pd
import os
import glob

def clean_trending_data():
    # 1. Load the JSON File
    # We look for the most recent JSON file in the data folder
    json_files = glob.glob('data/trends_*.json')
    if not json_files:
        print("No JSON files found in data/ folder. Please run Task 1 first.")
        return

    # Pick the latest file (sorted by name)
    input_file = sorted(json_files)[-1]

    # Load into a Pandas DataFrame
    df = pd.read_json(input_file)
    print(f"Loaded {len(df)} stories from {input_file}")

    # 2. Clean the Data

    # Duplicates: Remove rows with the same post_id
    df = df.drop_duplicates(subset=['post_id'])
    print(f"After removing duplicates: {len(df)}")

    # Missing values: Drop rows where critical fields are missing
    # subset=['post_id', 'title', 'score'] ensures we only drop if THESE are null
    df = df.dropna(subset=['post_id', 'title', 'score'])
    print(f"After removing nulls: {len(df)}")

    # Data types: Convert score and num_comments to integers
    df['score'] = df['score'].astype(int)
    df['num_comments'] = df['num_comments'].astype(int)

    # Low quality: Keep only stories with a score of 5 or higher
    df = df[df['score'] >= 5]
    print(f"After removing low scores: {len(df)}")

    # Whitespace: Clean up the title column
    # .str.strip() removes leading and trailing spaces
    df['title'] = df['title'].str.strip()

    # 3. Save as CSV
    output_path = 'data/trends_clean.csv'
    df.to_csv(output_path, index=False)

    print(f"\nSaved {len(df)} rows to {output_path}")

    # Print quick summary: Stories per category
    print("\nStories per category:")
    print(df['category'].value_counts())

# Run the cleaning script
if __name__ == "__main__":
    clean_trending_data()
```

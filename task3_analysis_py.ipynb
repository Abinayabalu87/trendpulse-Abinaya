```python
import pandas as pd
import numpy as np

def perform_analysis():
    # --- 1. LOAD THE CLEAN DATA ---
    try:
        # We load the clean CSV we created in Task 2
        df = pd.read_csv('data/trends_clean.csv')
        print(f"Loaded {len(df)} cleaned stories for analysis.")
    except FileNotFoundError:
        print("Error: 'data/trends_clean.csv' not found. Please run Task 2 first!")
        return

    # --- 2. CALCULATE STATISTICS (NumPy & Pandas) ---

    # We use .groupby() to split the data into categories
    # Then we use .agg() to calculate different things at once
    summary = df.groupby('category').agg({
        'score': ['mean', 'max'],           # Average score and Highest score
        'num_comments': ['mean', 'sum'],    # Average comments and Total comments
        'post_id': 'count'                  # How many stories in this category
    })

    # Cleaning up the column names so they are easier to read
    summary.columns = [
        'avg_score', 'max_score',
        'avg_comments', 'total_comments',
        'story_count'
    ]

    # Round the averages to 2 decimal places using NumPy
    summary['avg_score'] = np.round(summary['avg_score'], 2)
    summary['avg_comments'] = np.round(summary['avg_comments'], 2)

    # --- 3. FIND THE "TOP" CATEGORY ---
    # Using NumPy to find which category has the highest average score
    top_category = summary['avg_score'].idxmax()
    highest_avg = summary['avg_score'].max()

    print(f"\nAnalysis Complete!")
    print(f"The most popular category is '{top_category}' with an average score of {highest_avg}.")

    # --- 4. SAVE SUMMARY ---
    summary.to_csv('data/category_summary.csv')
    print("\nSummary saved to 'data/category_summary.csv'")

    # Display the summary table in the console
    print("\n--- Category Summary Table ---")
    print(summary)

if __name__ == "__main__":
    perform_analysis()
```
